```python
from PIL import Image
import numpy as np

images = [
    "dataset1.jpeg",
    "dataset2.jpeg",
    "dataset3.jpeg",
    "dataset4.jpeg",
    "dataset5.jpeg"
]

for img_name in images:

    img = Image.open(img_name)

    # convert to grayscale
    img = img.convert("L")

    # resize to 16x16
    img = img.resize((16,16))

    pixels = np.array(img)

    # flatten to 256 values
    pixels = pixels.flatten()

    mem_name = img_name.split(".")[0] + ".mem"

    with open(mem_name,"w") as f:
        for p in pixels:
            f.write("{:02x}\n".format(p))

    print(mem_name,"generated with",len(pixels),"values")
```
